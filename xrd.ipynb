```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction import xrd
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator as Abs
import matplotlib.pyplot as plt
#from pymatgen.analysis.diffraction.core import get_unique_families


with MPRester("nfHDloqMhx9N7jWGsB8") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-762")
    
    # Including data into a cif file
    w = CifWriter(structure)
    w.write_file('mp-762.cif')
    
        
    # On obtient le motif de diffraction    
    Difr = xrd.XRDCalculator(wavelength='CuKa')
    X = Difr.get_pattern(structure)   
    #XX = get_unique_families([1,1,0])
```

```python
print("Le premier pic a une intensitée de",X.y[0],"mise à l'échelle et un angle de diffraction 2\u0398 =",X.x[0],"degrés. Ses indices hkl associés sont:")
print(X.hkls[0][0]['hkl'],"avec une multiplicitée de",X.hkls[0][0]['multiplicity'])
print("la distance interplanaire associée à ce vecteur vaut d =",X.d_hkls[0])
```
